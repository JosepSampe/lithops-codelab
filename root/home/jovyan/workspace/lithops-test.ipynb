# Hybrid cloud Lithops example

## Lithops on Localhost node

```python
from lithops import FunctionExecutor

def hello(name):
    return 'Hello {}!'.format(name)

with FunctionExecutor(backend='localhost', storage='localhost') as fexec:
    fut = fexec.call_async(hello, 'World')
    print(fut.result())
```

## Lithops on IBM Cloud Functions and IBM Cloud Object Storage

```python
from lithops import FunctionExecutor

def mult(x):
    return x*5

with FunctionExecutor(backend='ibm_cf', storage='ibm_cos') as fexec:
    fexec.map(mult, range(4))
    print(fexec.get_result())
```

## Lithops on Local kubernetes cluster and IBM Cloud Object Storage

```python
from lithops import FunctionExecutor

def summ(x):
    return x+5

with FunctionExecutor(backend='k8s', storage='ibm_cos') as fexec:
    fexec.map(summ, range(4))
    print(fexec.get_result())
```
